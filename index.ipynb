Data Preprocessing

```python
data.head()
```

```python
data=data.drop(['Full Name'],axis=1)
```

```python
data.head()
```

Encoding the Categorical Columns

```python
data['Gender'].unique()
```

```python
data['Gender']=data['Gender'].map({'Male':0,'Female':1})
```

```python
data.head()
```

```python
data=data.drop(['Date of Birth','Preferred Learning Style','Parental Education Level','Monthly Household Income'],axis=1)
```

```python
data.head()
```

```python
data['University_Name'].unique()
```

```python
data['University_Name']=data['University_Name'].map({'University of Jaffna':0,'University of Kelaniya':1,'Rajarata University':2,'University of Moratuwa':3,'University of Colombo':4,'Eastern University':5,'University of Ruhuna':6,'University of Peradeniya':7,'Sabaragamuwa University':8})
```

```python
data.head()
```

```python
data['Current Academic Year'].unique()
```

```python
data['Current Academic Year']=data['Current Academic Year'].map({'1st':0,'2nd':1,'3rd':2,'4th':3})
```

```python
data.head()
```

```python
data['Participation in Extracurricular Activities'].unique()
```

```python
data['Participation in Extracurricular Activities']=data['Participation in Extracurricular Activities'].map({'No':0,'Yes':1})
```

```python
data.head()
```

Store Feature Matrix in X and Y(Target variable)

```python
data.columns
```

```python
y=data["Cumulative Grade Point Average (GPA)"]
X=data.drop(['Cumulative Grade Point Average (GPA)'],axis=1)
```

Splitting Dataset Into the Training Set and Test Set

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
data.head()
```

Import The Model

```python
threshold = 2.5
y_train_binary = (y_train > threshold).astype(int) 
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm 
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
threshold = 2.5
y_train_binary = (y_train > threshold).astype(int) 
```

```python
lr=LogisticRegression()
lr.fit(X_train,y_train_binary)
```

```python
svm=svm.SVC()
svm.fit(X_train,y_train_binary)
```

```python
knn=KNeighborsClassifier()
knn.fit(X_train,y_train_binary)
```

```python
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train_binary)
```

```python

rf=RandomForestClassifier()
rf.fit(X_train,y_train_binary)
```

```python

gb=GradientBoostingClassifier()
gb.fit(X_train,y_train_binary)
```

Prediction on Test Data

```python
y_pred1=lr.predict(X_test)
y_pred2=svm.predict(X_test)
y_pred3=knn.predict(X_test)
y_pred4=dt.predict(X_test)
y_pred5=rf.predict(X_test)
y_pred6=gb.predict(X_test)
```

Evaluating the Algorithms

```python
from sklearn.metrics import accuracy_score
```

```python
# Example of converting continuous predictions into binary class labels
threshold = 0.5  # Define a threshold to convert probabilities into binary predictions

y_pred_binary1 = (y_pred1 > threshold).astype(int)  # Convert to binary labels for Logistic Regression
y_pred_binary2 = (y_pred2 > threshold).astype(int)  # Convert to binary labels for SVM
y_pred_binary3 = (y_pred3 > threshold).astype(int)  # Convert to binary labels for KNN
y_pred_binary4 = (y_pred4 > threshold).astype(int)  # Convert to binary labels for Decision Tree
y_pred_binary5 = (y_pred5 > threshold).astype(int)  # Convert to binary labels for Random Forest
y_pred_binary6 = (y_pred6 > threshold).astype(int)  # Convert to binary labels for Gradient Boosting
```

```python
y_test_binary = (y_test > threshold).astype(int)  # Convert to binary labels
```

```python
score1=accuracy_score(y_test_binary,y_pred_binary1)
score2=accuracy_score(y_test_binary,y_pred_binary2)
score3=accuracy_score(y_test_binary,y_pred_binary3)
score4=accuracy_score(y_test_binary,y_pred_binary4)
score5=accuracy_score(y_test_binary,y_pred_binary5)
score6=accuracy_score(y_test_binary,y_pred_binary6)
```

```python
print(score1,score2,score3,score4,score5,score6)
```

```python
final_data=pd.DataFrame({'Models':['LogisticRegression','SuportedVectorMachine','KNeighborsClassifier','DecisionTreeClassifier','RandomForestClassifier','GradientBoostingClassifier'],
                         'Accuracy':[score1*100,
                                score2*100,
                                score3*100,
                                score4*100,
                                score5*100,
                                score6*100]})
```
